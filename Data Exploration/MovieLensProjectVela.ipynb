### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
# creating new data frame , setting release year as index , dropping columns not necessary for analysis 
itemyeargenreindex= item.set_index('releaseyear')
itemyeargenreindex.drop(columns='release date',inplace = True)
itemyeargenreindex.drop(columns='movie title' ,inplace = True)
itemyeargenreindex.drop(columns='movie id' ,inplace = True)
itemyeargenreindex.drop(columns='unknown' ,inplace = True)
itemyeargenreindex.drop(columns='releasedate' ,inplace = True)
# grouping by release year and sum of all columns
itemyeargenregrouped=itemyeargenreindex.groupby('releaseyear').sum()
```

```python
# finding the no of releases per year adding a column for storing the data 
itemyeargenregrouped['totalcount']=item['releaseyear'].value_counts()
# finding percent of movies in a given year and storing in a new data frame
itemyeargenregroupedpercent=(itemyeargenregrouped.div(itemyeargenregrouped['totalcount'], axis= 0) * 100)
itemyeargenregroupedpercent.tail()
```

```python
# heatmap to visualise all the changes over the years 
plt.figure(figsize=(12, 8))
sns.heatmap(itemyeargenregroupedpercent)
plt.show()
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# creating a new dataframe and grouping by movie id and average rating
mymergeratingaverage=data.groupby('movie id').mean()['rating']

# creating a new dataframe and grouping by movie id and count of ratings
mymergeratingcount=data.groupby('movie id')['rating'].value_counts()

# merging average rating and count to a single data frame
mymergeds=pd.merge(mymergeratingaverage,mymergeratingcount,how='outer',on='movie id')
# renaming columns
mymergeds.rename(columns = {'rating_x':'average_rating','rating_y':'count_rating'},inplace=True)
# merging with item dataframe to get the movie name
mymergedsnames=pd.merge(mymergeds,item[['movie id', 'movie title']],how='outer',on='movie id')

#Find the count of ratings and average ratings for every movie.
mymergedsnames.head()
```

```python
# Slice the movies which have ratings more than 100.
mymergedsnames[mymergedsnames['count_rating'] > 100]
```

```python
# Sort values according to average rating such that movie which highest rating is on top.
mymergedsnames.sort_values(by='average_rating', ascending=False)
```

```python
# Select top 25 movies
mymergedsnames.sort_values(by='average_rating', ascending=False).head(25)
```
